```python
%matplotlib inline

import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
```

```python
dados = stats.gamma.rvs(scale=10,loc=4, a = 6, size=100000)
MeanDados = np.mean(dados)
VarDados = np.var(dados)
DPDados = VarDados**(1/2)
```

```python
n = 2
random = np.random.choice(dados,n)
MeanRand = 0
for i in range(len(random)):
    MeanRand += random[i]
MeanRand = MeanRand/n
```

```python
ListMean = list()

for i in range(100000):
    MeanRand = 0
    random = np.random.choice(dados,n)
    for j in range(len(random)):
        MeanRand += random[j]
    ListMean.append(MeanRand/n)
```

```python
plt.hist(dados, bins = 20, normed = 1)
plt.axis([0,200,0,0.018])
plt.grid(True)
plt.show()
```

```python
x = DPDados/((n**(1/2)))

t = np.linspace(0,200,4000)
plt.hist(ListMean, bins = 20, normed = True, color='gold')
plt.grid(True)
plt.title("Para 10000 simulações")
plt.ylabel("Densidade")
plt.xlabel("Médias Amostrais")
plt.plot(t,stats.norm.pdf(t,scale=x,loc=MeanDados), color = 'red')
plt.show()
```

```python
n = 50
x = DPDados/((n**(1/2)))

listaX1 = list()
listaX9 = list()

random = np.random.choice(dados,n)
MeanRand = 0
for i in range(len(random)):
    MeanRand += random[i]
MeanRand = MeanRand/n

ListMean = list()

for i in range(100000):
    MeanRand = 0
    random = np.random.choice(dados,n)
    for j in range(len(random)):
        MeanRand += random[j]
        
    listaX1.append(random[0])
    listaX9.append(random[8])
    
    ListMean.append(MeanRand/n)
    
t = np.linspace(0,200,4000)
plt.hist(ListMean, bins = 20, normed = True, color='gold')
plt.grid(True)
plt.title("Para 10000 simulações")
plt.ylabel("Densidade")
plt.xlabel("Médias Amostrais")
plt.plot(t,stats.norm.pdf(t,scale=x,loc=MeanDados), color = 'red')
plt.show()
```
